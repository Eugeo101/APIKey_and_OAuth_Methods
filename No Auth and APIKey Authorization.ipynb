# ReCAP

```python
result = {
    "employees":
    [
        {
            "id":1,
            "name":"Ahmed",
            "Salary":32800,
            "age":24
        },
        {
            "id":2,
            "name":"Mohamed",
            "Salary":63500,
            "age":27
        },
        {
            "id":3,
            "name":"Sara",
            "Salary":58900,
            "age":66
        }
        
    ]
}
result
```

```python
type(result)
```

```python
import json

result_json = json.dumps(result) # convert data -> json_format to be sent
result_json
```

```python
type(result_json)
```

```python
result_converted = json.loads(result_json) # convert json_format -> python datatype
result_converted
```

```python
with open("employees.csv", mode='w', encoding='utf-8') as fd:
    fd.write("id,name,salary,age\n")
    for person in result_converted['employees']:
        p_id = person['id']
        name = person['name']
        salary = person['Salary']
        age = person['age']
        fd.write(f"{p_id},{name},{salary},{age}\n")
```

```python
import pandas as pd

df = pd.read_csv("employees.csv")
df
```
